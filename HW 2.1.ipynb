```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
```

```python
dfs = pd.read_html(url, match="2012")
df = dfs[2]
```

```python
df.shape
```

```python
df = df.replace("—", np.nan)
df
```

```python
df.dtypes
```

```python
# Перетворюю всі значення на тип float, та зміщаю кому на 1 вліво, бо воно неправильно зчитало та значення (23,0 записало як 230.0)
df[['1950', '1960', '1970', '1990', '2000', '2012', '2014',
       '2019']] = df[['1950', '1960', '1970', '1990', '2000', '2012', '2014',
       '2019']].astype(float) * 0.1
df
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—":
for col in df.iloc[:,1:]:
  if df[col].dtype != 'float64':
    print(df[col].dtype)
    df[col] = df[col].astype(float)
df.dtypes
```

```python
df.isna().sum()
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
df = df.drop([len(df)-1])
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
# можливо є кращий спосіб це зробити?
df.iloc[:,1:] = df.iloc[:,1:].fillna(df.iloc[:,1:].mean())
df = df.round(decimals = 1)
df
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
df['2019'].mean()
```

```python
df[df['2019'] > df['2019'].mean()][["Регіон", "2019"]]
```

```python
# У якому регіоні була найвища народжуваність у 2014 році
df[df['2014'] == df['2014'].max()]['Регіон']
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
# Створимо датафрейм для побудови графіка, щоб регіони були індексами (думаю, це потрібно було зробити раніше❓😂):
df_region_2019 = df[['Регіон', '2019']].set_index('Регіон')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize = (20, 20))
ax = df_region_2019.plot(kind = 'bar', title = "Коефіцієнт народжуваності по регіонах за 2019 рік")

ax.set_xlabel("Регіон")
ax.set_ylabel("Народжуваність")

plt.show()
```

```python
# тепер працює(раніше через "Язык.программирования" не працювало)
python_data = df.query('Язык_программирования == "Python"')
```

```python
python_data.shape
```

```python
python_data.columns
```

```python
## Можливо тут якось по іншому потрібно було?🙃
job_title_data = pd.DataFrame(python_data.groupby(['Должность'])\
                              .agg(max_salary=('Зарплата_в_месяц', max), min_salary=('Зарплата_в_месяц', 'min')))
```

```python
job_title_data.head(10)
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. 
# Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    return (row['max_salary'] + row['min_salary']) / 2
```

```python
job_title_data['avg'] = job_title_data.apply(fill_avg_salary, axis=1)
```

```python
# —Ç–µ–ø–µ—Ä –ø—Ä–∞—Ü—é—î(—Ä–∞–Ω—ñ—à–µ —á–µ—Ä–µ–∑ "–Ø–∑—ã–∫.–ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è" –Ω–µ –ø—Ä–∞—Ü—é–≤–∞–ª–æ)
python_data = df.query('–Ø–∑—ã–∫_–ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è == "Python"')
```

```python
python_data.shape
```

```python
python_data.columns
```

```python
## –ú–æ–∂–ª–∏–≤–æ —Ç—É—Ç —è–∫–æ—Å—å –ø–æ —ñ–Ω—à–æ–º—É –ø–æ—Ç—Ä—ñ–±–Ω–æ –±—É–ª–æ?üôÉ
job_title_data = pd.DataFrame(python_data.groupby(['–î–æ–ª–∂–Ω–æ—Å—Ç—å'])\
                              .agg(max_salary=('–ó–∞—Ä–ø–ª–∞—Ç–∞_–≤_–º–µ—Å—è—Ü', max), min_salary=('–ó–∞—Ä–ø–ª–∞—Ç–∞_–≤_–º–µ—Å—è—Ü', 'min')))
```

```python
job_title_data.head(10)
```

```python
# –°—Ç–≤–æ—Ä—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é fill_avg_salary, —è–∫–∞ –ø–æ–≤–µ—Ä—Ç–∞—Ç–∏–º–µ —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –∑–∞—Ä–æ–±—ñ—Ç–Ω–æ—ó –ø–ª–∞—Ç–∏ –Ω–∞ –º—ñ—Å—è—Ü—å. 
# –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —ó—ó –¥–ª—è –º–µ—Ç–æ–¥—É apply —Ç–∞ —Å—Ç–≤–æ—Ä—ñ—Ç—å –Ω–æ–≤–∏–π —Å—Ç–æ–≤–ø—á–∏–∫ "avg"
def fill_avg_salary(row):
    return (row['max_salary'] + row['min_salary']) / 2
```

```python
job_title_data['avg'] = job_title_data.apply(fill_avg_salary, axis=1)
```
